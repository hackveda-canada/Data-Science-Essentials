```
#Extract text from audio files
import textract
text = str(textract.process("audio1.mp3"))
text
```

```
#word Tokenization using NLTK Library
#https://en.wikipedia.org/wiki/Lexical_analysis#Tokenization

#Extract sentences from text: Sentence Tokenization
#Extract words from text: Word Tokenization

import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize

tokens = word_tokenize(text)
tokens

```

```
# Sentence Tokenization
!wget https://newsroom.intel.com/wp-content/uploads/sites/11/2018/03/smart-cities-whats-in-it-for-citizens.pdf
```

```
textpdf = textract.process("smart-cities-whats-in-it-for-citizens.pdf")
textpdf
```

```
# Perform Sentence Tokenization
from nltk.tokenize import sent_tokenize
sentences = sent_tokenize(str(textpdf))
sentences

counter = 1

#Print each sentence in sentences
for sentence in sentences:
  print("sentence ",str(counter), sentence)
  counter+=1
```

```
# We can also perform tokenization using TreebankWordTokenizer
from nltk.tokenize import TreebankWordTokenizer

tokenizer = TreebankWordTokenizer()
tokenizer.tokenize(str(textpdf))
#tokenizer.tokenize_sents(str(textpdf))
```
