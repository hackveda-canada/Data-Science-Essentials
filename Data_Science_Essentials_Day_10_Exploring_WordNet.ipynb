```
#Explore WordNet
#WordNet = Lexical Database

import nltk
nltk.download("wordnet")
from nltk.corpus import wordnet as wn

#WordNet contains : Nouns, Adjectives, Verbs, Adverbs and Synonyms/Antonyms
#Identify synonyms (lemma) out of a wordform in Natural Language Text
#Word -> Synonyms , WordNet(Synsets)

```

```
#Synonyms for the word
synsets = wn.synsets("left")

#Number of synonyms
print("The number of synonym sets", len(synsets))
```

```
#Explore each set of synonym for word
for synset in synsets :
  #Synset contain Part Of Speech : (Noun, Adjective, Verb, Adverb),
  #Definition, Lemmas, Examples
  print(synset)
  print(synset.lexname(), "(", synset.pos(), ")")  #To find POS
  print("Definition: ", synset.definition(), "\n")
  print("Lemma Words: ", synset.lemma_names())
  print("Examples: ", synset.examples())
  print("\n")
```
